```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
# Generating random data for 1000 values
np.random.seed(0)
data = {'embed_0': np.random.rand(1000),
        'embed_767': np.random.rand(1000)}
df = pd.DataFrame(data)
 
# Scatter plot using seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='embed_0', y='embed_767', data=df, alpha=0.6)
 
# Setting title and labels
plt.title('Scatter Plot of embed_0 vs embed_767')
plt.xlabel('embed_0')
plt.ylabel('embed_767')
 
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
file_path = r"C:\Users\DELL\Downloads\t5_train (1).xlsx"
 
# Reading data from the Excel file into a DataFrame
df = pd.read_excel(file_path)
 
 
# Scatter plot using seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='embed_0', y='embed_767', data=df, alpha=0.6)
 
# Setting title and labels
plt.title('Scatter Plot of embed_0 vs embed_767')
plt.xlabel('embed_0')
plt.ylabel('embed_767')
 
# Display the plot
plt.show()
 
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = df['embed_0'].values.reshape(-1, 1)
y = df['embed_767'].values
 
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
 
# Creating and training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
 
# Making predictions
y_pred = model.predict(X_test)
 
# Calculating mean squared error
mse = mean_squared_error(y_test, y_pred)
 
print("Mean Squared Error:", mse)
```
